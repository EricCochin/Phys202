#### 1 A 2 1 Dfinitions
- Une quation diffrentielle est une quation dans laquelle sont utilises des drives de fonctions de variables dpendantes par rapport  une ou plusieurs variables indpendantes (le temps, des repres de positions).
- Une quation diffrentielle est *ordinaire* si une seule variable indpendante est utilise, sinon c'est une quation aux *drives partielles*.
- Les oprations de drivations peuvent tre reprsents par 3 notations, *Leibniz*: $\ \newcommand{\deriv}{\mathrm{d}} \frac{\deriv z}{\deriv t }$, *prime*: $\ z'$ , et *Newton*: $\ \dot{z}$.
- L'*ordre* d'une quation diffrentielle est le degr le plus lev de drivation: $\ \frac{\deriv z}{\deriv t } = t^2$ est une quation diff. du premier ordre, $\ \ddot{y} +y\cdot \dot{x}^3 = 0 $ est une equation diff du deuxime ordre.
- La *forme gnrale* d'une quation diffrentielle est $\ F(t, x, \dot{x},\ddot{x}, ...) = 0 \tag{14}$ et sa *forme normale* ou *rsolue*est telle que $\ \frac{\deriv^n x}{\deriv t^n} = f(t, x,\frac{\deriv x}{\deriv t} ,...,\frac{\deriv^{n-1} x}{\deriv t^{n-1}} ) \tag{15}$, dans laquelle le terme de plus haut rang est isol.
- Le paramtre le plus important pour caractriser une quation diffrentielle est la distinction linaire vs non linaire. Une quation diff. linaire est une combinaison linaire de la variable dpendante et de ses drives, par exemple $\ a_2(t) \cdot \frac{\deriv^2 x}{\deriv t^2}+a_1(t) \cdot \frac{\deriv x}{\deriv t} + a_0(t) \cdot x = g(t) \tag{16}$ est une quation diff lineaire du deuxime ordre. Par contre  $\ a_1(t) \cdot \frac{\deriv x}{\deriv t} + a_0(t) \cdot x^2 = h(t) \tag{17}$ est une quation diff du premier ordre non linaire.

- Une caractristique importante de ce type d'quation est le *principe de superposition*, si $\ y_1 $ et $ y_2 $ sont des solutions distinctes de l'quation homogne $\ a \cdot \ddot{y} + b \cdot \dot{y} + c \cdot y = 0 \tag{23}$ alors $\ y_1 + y_2 \tag{24}$ est aussi une solution, ainsi que toutes les  combinaisons linaires: $\ \alpha y_1 + \beta y_2 \tag{25}$ o $\ (\alpha,\beta) \in \mathbb{R}^2$ 

- Par exemple $\ y_1=\cos(t)$ et $\ y_2 = \sin(t) $ sont deux solutions possibles de l'quation $\ \ddot{y} - y = 0$, et $\ y_1 + y_2 $ ou $\ 5 y_1 - 3 y_2 $ sont aussi des solutions de cette quation.

- Pour rsoudre une quation inhomognes on combine les solutions de l'quation homogne et une seule solution de l'quation inhomogne. Par exemple $\ y_1=\cos(t) \tag{26}$ et $\ y_2 = \sin(t) \tag{24}$ sont des solutions de $\ \ddot{y} - y = 0 \tag{27}$ et $\ y_p = 1 \tag{28}$ est une solution de $\ \ddot{y} - y = 1 \tag{29}$ Les solutions de (20)  sont donc  $\ \alpha y_1 + \beta y_2  + y_p \tag{30}$ o $\ (\alpha,\beta) \in \mathbb{R}^2$.

### 1 A 3 1 Mcanique
Les problmes de mcanique commencent avant mme de modliser un objet par le choix d'un **rfrentiel**. On considrera qu'un rfrentiel li au laboratoire a les proprits d'un rfrenciel Galilen. Cette hypothse est une approximation qu'il convient de discuter si le problme tudi est la vibration d'une fuse pendant un voyage spatial ou le comportement d'un simple pendule pendant plusieurs heures (pendule de Foucault).

- Rfrentiel : Constitu d'un point fixe, et deux ou trois axes dots de vecteurs directeurs. Not R ou (Oxyz), les vecteurs directeurs sont nots $\ \overrightarrow{e}_x, \overrightarrow{e}_y, \overrightarrow{e}_z $ forment un tridre direct $\ \overrightarrow{e}_x \land \overrightarrow{e}_y = \overrightarrow{e}_z$, o le symbole $\ \land $ est le produit vectoriel. 

Rappel : le produit vectoriel de deux vecteurs, $\ \overrightarrow{A} \land \overrightarrow{B}$, est donn par $\ \begin{pmatrix}A_1 \\ A_2 \\ A_3 \end{pmatrix} \land \begin{pmatrix}B_1 \\ B_2 \\ B_3 \end{pmatrix} = \begin{pmatrix}A_2B_3 -A_3B_2 \\ -A_1B_3+A_3B_1 \\ A_1B_2-A_2B_1 \end{pmatrix}$

La mcanique est la description du mouvement des objets, avant de dcrire des mouvements prcisont que les objets tudis dans ce cours sont modlis ou par un point matriel (un objet dont la dimension n'est pas prise en compte, une masse quasi ponctuelle, pas de rotation sur lui mme) ou un objet solide (un objet indformable, pouvant cumuler translation et rotation).

- **Lois de Newton**: 3 lois permettent pour des objets simples d'obtenir les quations du mouvement (l'objet a une masse m constante, rfrentiel galilen)

    - Troisime loi: Tout objet subissant une force extrieure exerce sur la source de celle ci une force gale mais oppose en direction.

La troisime lois est utile pour dcrire les interactions entre systme, mais c'est la deuxime qui permet d'obtenir l'quation vectorielle du mouvement.
