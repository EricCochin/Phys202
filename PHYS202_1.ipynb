### 1 A 1 Nombres complexes
Les nombres complexes ont été introduit comme une astuce mathématique pour résoudre les équations algébriques du type $\ x^2-1=0$. Aucune quantité Physique ne peut être décrite par un nombre complexe, pourtant il est fait un large usage dans le domaine des vibrations ou en electronique. La raison en est que dans ces domaines les quantités mesurables sont  caractérisées par deux paramètres, l'un étant une amplitude (de mouvement, de tension, de force, ...) et le deuxième un paramètre de phase (pour repérer le mouvement, la tension, la perturbation extérieure). Ces deux informations peuvent être d'un point de vue pratique codées en utilisant une seule quantité complexe lors des calculs. Par contre, lorsque lors de l'interprétation ou présentation des résultats, il ne faut considérer que les quantités réelles, observables, mesurables.

### 1 A 1 a Définitions
Un nombre complexe z est la combinaison de deux nombres réels a et b, il s'écrit sous sa *forme algébrique*
$\ z = a + j b$ ou j est tel que $\ j^2 = -1$. La notation j plutôt que i pour le nombre imaginaire pur est utilisée car dans les problèmes électroniques i fait souvent référence à l'intensité du courant électrique. Les opérations arithmétiques usuelles sont élargies aux quantités complexes:
$\ z_1 = a_1 + j b_1 et z_2 = a_2 + j b_2 \tag{1}$

$\ z_1 + z_2 = (a_1 + a_2) + j (b_1 + b_2) \tag{2}$

$\ z_1 - z_2 = (a_1 - a_2) + j (b_1 - b_2) \tag{3}$

$\ z_1 * z_2 = (a_1 * a_2 - b_1 * b_2) + j (a_1 * b_2 + a_2 * b_1) \tag{4}$

$\ \frac{z_1}{z_2} = \frac{a_1 + j b_1}{a_2 + j b_2} = \frac{(a_1 + j b_1)(a_2 - j b_2)}{a_2^2+b_2^2} \tag{5}$

On appelle $\ z^*$ le conjugué de $\ z = a + j b$, tel que $\ z^* = a - j b \tag{6}$

La forme algébrique des nombres complexes est utile mais une autre représentation est souvent utilisée, la *forme géométrique*. Considérant que le nombre z = a  + j b est un point de coordonnées (a , b) dans un plan dit complexe ou diagramme d'Argand, dans ce plan complexe l'axe horizontal est l'axe réel, l'axe vertical est l'axe imaginaire. Par exemple dans la figure 2 sont représentés les nombres complexes A = 3 + 2j, B = -2 + 4 j, C = -3 - 5 j et D = 1 - 3 j.



*Figure 2: Diagramme d'Argand*


La représentation d'un nombre complexe dans le digramme d'Argand, invite à une autre représentation, la *forme trigonométrique* des nombres complexes. Dans la figure 3 est représenté le nombre z = x +j y, le point de coordonnées (x, y) peut être repéré par la distance r et un angle $\ \theta$, r est le *module* de z et l'angle $\ \theta$ est l'*argument* de z.



*Figure 3: Forme trigonométrique.*

Le module de z est souvent noté $\ \left | z \right |$ est égal à $\ \sqrt{x^2 + y^2} \tag{7}$ et son argument $\ \theta$ est noté $\ \arg(z)$ que $\ \arg(z) = \theta =\tan^{-1}(\frac{y}{x}) \tag{8}$. Lors de l'évaluation numérique il faut s'assurer que le résultat est bien l'argument car la fonction inverse de tangente ne produit que des valeurs dans l'intervalle $\ \left[-\frac{\pi}{2},\frac{\pi}{2} \right] $ alors que l'argument est définit dans l'intervalle $\ \left[-\pi,\pi \right] $


.
*Figure 4: Argument de z = a + j b*

Soit si $\ a>0 \Rightarrow \arg(z) = \tan^1(\frac{b}{a})$
    si $\ a<0 \: et \: b>0 \Rightarrow \arg(z) = \tan^1(\frac{b}{a})+\pi$ et si $\ a<0 \: et \: b<0 \Rightarrow \arg(z) = \tan^1(\frac{b}{a})-\pi$. La forme trigonométrique est bien adaptée aux opérations de multiplications (ou divisions), $\ z_1*z_2 = r_1*r_2*\left(\cos(\theta_1 + \theta_2) + j \sin(\theta_1 + \theta_2) \right) \tag{9}$ et $\ \frac{z_1}{z_2} = \frac{r_1}{r_2}*\left(\cos(\theta_1 - \theta_2) + j \sin(\theta_1 - \theta_2) \right) \tag{10}$.


Il existe une troisième représentation des nombres complexes, la *forme exponentielle*. Cette forme est fondée sur la formule d'Euler $\ e^{j x} = \cos(x) + j \sin(x) \tag{11}$, le nombre complexe $\ z = \left|z\right| \left( \cos(\theta) + j \sin(\theta) \right) \tag{12}$ peut être représenté par $z = \left|z\right| e^{j\theta}$; Cette dernière forme est adaptée aux opérations de multiplications ou de mise à la puissance, en effet $\ \left( r e^{j \theta}\right)^n =  r^n e^{j n \theta} \tag{13}$.

### 1 A 2 Equations différentielles

#### 1 A 2 1 Définitions
- Une équation différentielle est une équation dans laquelle sont utilisées des dérivées de fonctions de variables dépendantes par rapport à une ou plusieurs variables indépendantes (le temps, des repères de positions).
- Une équation différentielle est *ordinaire* si une seule variable indépendante est utilisée, sinon c'est une équation aux *dérivées partielles*.
- Les opérations de dérivations peuvent être représentés par 3 notations, *Leibniz*: $\ \newcommand{\deriv}{\mathrm{d}} \frac{\deriv z}{\deriv t }$, *prime*: $\ z'$ , et *Newton*: $\ \dot{z}$.
- L'*ordre* d'une équation différentielle est le degré le plus élevé de dérivation: $\ \frac{\deriv z}{\deriv t } = t^2$ est une équation diff. du premier ordre, $\ \ddot{y} +y\cdot \dot{x}^3 = 0 $ est une equation diff du deuxième ordre.
- La *forme générale* d'une équation différentielle est $\ F(t, x, \dot{x},\ddot{x}, ...) = 0 \tag{14}$ et sa *forme normale* ou *résolue*est telle que $\ \frac{\deriv^n x}{\deriv t^n} = f(t, x,\frac{\deriv x}{\deriv t} ,...,\frac{\deriv^{n-1} x}{\deriv t^{n-1}} ) \tag{15}$, dans laquelle le terme de plus haut rang est isolé.
- Le paramètre le plus important pour caractériser une équation différentielle est la distinction linéaire vs non linéaire. Une équation diff. linéaire est une combinaison linéaire de la variable dépendante et de ses dérivées, par exemple $\ a_2(t) \cdot \frac{\deriv^2 x}{\deriv t^2}+a_1(t) \cdot \frac{\deriv x}{\deriv t} + a_0(t) \cdot x = g(t) \tag{16}$ est une équation diff lineaire du deuxième ordre. Par contre  $\ a_1(t) \cdot \frac{\deriv x}{\deriv t} + a_0(t) \cdot x^2 = h(t) \tag{17}$ est une équation diff du premier ordre non linéaire.
- La (les) solution(s) d'une équation différentielle peuvent être illustrées graphiquement. Par exemple le graphe ci dessous représente quelques solutions de $\ \dot{y} = -\frac{y}{10} +\cos(t) \tag{18}$ à partir de différentes conditions initiales.

- Dernier point de vocabulaire, une équation différentielle peut être homogène:  $\ F(t, x, \dot{x},\ddot{x}, ...) = 0 \tag{20}$ ou inhomogène: $\ F(t, x, \dot{x},\ddot{x}, ...) = f(t) \tag{21}$. Ce point sera souvent utilisé car un système sera modélisé par une équation homogène après l'extinction d'une perturbation (pas de forces extérieures, générateurs de tension et de courant éteint) et par une équation inhomogène lorsqu'une perturbation extérieure s'exerce (force extérieure, générateur de tension en fonction, ...).

#### 1 A 2 2 Equations différentielles linéaires du second ordre à coefficients constants
Dans ce module, les phénomènes rencontrés seront modélisés par un sous ensemble extrémement limité d'équations différentielles, en l'occurence des équations du second ordres, lineaires, à coefficients constants:$\ a \cdot \ddot{y} + b \cdot \dot{y} + c \cdot y = f(t) \tag{22}$

- Une caractéristique importante de ce type d'équation est le *principe de superposition*, si $\ y_1 $ et $ y_2 $ sont des solutions distinctes de l'équation homogène $\ a \cdot \ddot{y} + b \cdot \dot{y} + c \cdot y = 0 \tag{23}$ alors $\ y_1 + y_2 \tag{24}$ est aussi une solution, ainsi que toutes les  combinaisons linéaires: $\ \alpha y_1 + \beta y_2 \tag{25}$ où $\ (\alpha,\beta) \in \mathbb{R}^2$ 

- Par exemple $\ y_1=\cos(t)$ et $\ y_2 = \sin(t) $ sont deux solutions possibles de l'équation $\ \ddot{y} - y = 0$, et $\ y_1 + y_2 $ ou $\ 5 y_1 - 3 y_2 $ sont aussi des solutions de cette équation.
- Par contre ce principe n'est plus applicable si l'équation est inhomogène:$\ y_1=\cos(t) -1$ et $\ y_2 = \sin(t)-1 $ sont deux solutions possibles de l'équation $\ \ddot{y} - y = 1$, alors que  $\ y_1 + y_2 $ ou $\ 5 y_1 - 3 y_2 $ ne sont pas des solutions de cette équation. 
- Pour résoudre une équation inhomogènes on combine les solutions de l'équation homogène et une seule solution de l'équation inhomogène. Par exemple $\ y_1=\cos(t) \tag{26}$ et $\ y_2 = \sin(t) \tag{24}$ sont des solutions de $\ \ddot{y} - y = 0 \tag{27}$ et $\ y_p = 1 \tag{28}$ est une solution de $\ \ddot{y} - y = 1 \tag{29}$ Les solutions de (20)  sont donc  $\ \alpha y_1 + \beta y_2  + y_p \tag{30}$ où $\ (\alpha,\beta) \in \mathbb{R}^2$. 
- Pour cette dernière solution, il apparait deux coefficients à priori, quelconques, ces derniers sont fixés par des informations supplémentaires, des conditions initales en général.
- Les solutions d'une équation différentielle homogène à coefficients constants sont des fonctions exponentielles $\ e^{s t} \tag{31}$ où $\ s \in \mathbb{C}$. 
- Les équations différentielles rencontrés dans ce cours seront toujours des équations différentielles à coefficients constants homogènes ou inhomogènes. Les fonctions exponentielles $\ e^{s t}$ sont la base des solutions à tous les problèmes rencontrés ultérieurement et modélisés par des équations homogènes.  
- Si une fonction exponentielle $\ y_s = e^{s t}$ est utilisée dans l'équation (23), on obtient l'équation $\ (a \cdot s^2 + b \cdot{s} + c )\cdot y_s = 0 \tag{32}$. 
- L'équation $\ a \cdot s^2 + b \cdot{s} + c  = 0 \tag{33}$ est l'*équation caractéristique* de l'équation différentielle $\ a \cdot \ddot{y} + b \cdot \dot{y} + c \cdot y = 0 \tag{34}$.
- L'équation caractéristique fixe deux valeurs possibles pour le coefficient s dans les solutions $\ e^{s t}$.
- Le discriminant $\ \Delta$ de l'équation caractéristique fixe les propriétés essentielles des solutions.
- Le discriminant: 
$\ \begin{equation} \Delta = (b^2 - 4ac) \tag{35} \end{equation}$ 

peut être positif, négatif ou nul selon les valeurs des paramètres a, b et c rencontrés, ces trois cas sont acceptables mais correspondent à des régimes distincts.


--------
- Si le discriminant est **positif**, l'équation caractéristique admet deux solutions réelles, distinctes et positives: 
$\ \left\{ \begin{array}{r l c}
 s_1 & = & \frac{-b - \sqrt{b^2 - 4ac}}{2 a} \\
 s_2 & = & \frac{-b + \sqrt{b^2 - 4ac}}{2 a} \end{array} \right. \tag{36} $
 
    - Il y a donc deux solutions exponentielles distinctes:
$\ \left\{ \begin{array}{r l c}
 y_1  & = & e^{s_1 t}\\
 y_2 & = & e^{s_2 t} \end{array} \right. \tag{37} $
    - Une solution réelle de l'équation homogène (34) est une combinaison de ces deux exponentielles: 
 $\ y = \alpha e^{s_1 t} + \beta e^{s_2 t} \tag{38}$ où $\ (\alpha,\beta) \in \mathbb{R}^2$
    - les coefficients $\ \alpha$ et $\ \beta$ sont fixés par les conditions initiales.
    
--------

--------
- Si le discriminant est **négatif**, l'équation caractéristique admet deux solutions complexes, distinctes et positives: 
$\ \left\{ \begin{array}{r l c}
 s_1 & = & \frac{-b - j\sqrt{4ac-b^2}}{2 a} \\
 s_2 & = & \frac{-b + j\sqrt{4ac-b^2}}{2 a} \end{array} \right. \tag{39} $
    - Il y a donc deux solutions exponentielles distinctes:
$\ \left\{ \begin{array}{r l c}
 y_1  & = & e^{s_1 t}\\
 y_2 & = & e^{s_2 t} \end{array} \right. \tag{40} $
    - Une solution réelle de l'équation homogène (34) est une combinaison de ces deux exponentielles: 
 $\ y = \alpha e^{s_1 t} + \beta e^{s_2 t} \tag{41}$ où $\ (\alpha,\beta) \in \mathbb{C}^2 $ et $\ \beta = \alpha^*$
    - les coefficients $\ \alpha$ et $\ \beta$ sont fixés par les conditions initiales.
    
--------

--------
- Si le discriminant est **nul**, l'équation caractéristique admet une seule solution réelle: 
$\ s_0  =  \frac{-b }{2 a}  \tag{42} $
 
    - Une solution réelle de l'équation homogène (34) est la combinaison : 
 $\ y = \left( \alpha t  + \beta \right) e^{s_0 t} \tag{43}$ où $\ (\alpha,\beta) \in \mathbb{R}^2 $ 
    - les coefficients $\ \alpha$ et $\ \beta$ sont fixés par les conditions initiales.

### 1 A 3 1 Mécanique
Les problèmes de mécanique commencent avant même de modéliser un objet par le choix d'un **référentiel**. On considérera qu'un référentiel lié au laboratoire a les propriétés d'un référenciel Galiléen. Cette hypothèse est une approximation qu'il convient de discuter si le problème étudié est la vibration d'une fusée pendant un voyage spatial ou le comportement d'un simple pendule pendant plusieurs heures (pendule de Foucault).

- Référentiel : Constitué d'un point fixe, et deux ou trois axes dotés de vecteurs directeurs. Noté R ou (Oxyz), les vecteurs directeurs sont notés $\ \overrightarrow{e}_x, \overrightarrow{e}_y, \overrightarrow{e}_z $ forment un trièdre direct $\ \overrightarrow{e}_x \land \overrightarrow{e}_y = \overrightarrow{e}_z$, où le symbole $\ \land $ est le produit vectoriel. 

Rappel : le produit vectoriel de deux vecteurs, $\ \overrightarrow{A} \land \overrightarrow{B}$, est donné par $\ \begin{pmatrix}A_1 \\ A_2 \\ A_3 \end{pmatrix} \land \begin{pmatrix}B_1 \\ B_2 \\ B_3 \end{pmatrix} = \begin{pmatrix}A_2B_3 -A_3B_2 \\ -A_1B_3+A_3B_1 \\ A_1B_2-A_2B_1 \end{pmatrix}$

La mécanique est la description du mouvement des objets, avant de décrire des mouvements précisont que les objets étudiés dans ce cours sont modélisé ou par un point matériel (un objet dont la dimension n'est pas prise en compte, une masse quasi ponctuelle, pas de rotation sur lui même) ou un objet solide (un objet indéformable, pouvant cumuler translation et rotation).

---------------
- Mouvements de translation:
Le mouvement d'un objet solide en l'absence de rotation est régis par une seule équation vectorielle du mouvement. Cette équation est accessibles par plusieurs méthodes, application des lois de Newton, application du théorème de l'énergie mécanique, par contre les méthodes de Lagrange ou d'Hamilton ne seront pas utilisées. 
---------------

- **Lois de Newton**: 3 lois permettent pour des objets simples d'obtenir les équations du mouvement (l'objet a une masse m constante, référentiel galiléen)

    - Première loi: Un objet soumis à aucune force ou à des forces qui s'annullent est en *mouvement uniforme*.
    - Deuxième loi: La somme des forces extérieures qui s'exercent sur un objet de masse m est égale à sa masse multipliée par son accélération: $\ m \overrightarrow\gamma = \sum{\overrightarrow{F}_{ext}} \tag{43}$
    - Troisième loi: Tout objet subissant une force extérieure exerce sur la source de celle ci une force égale mais opposée en direction.

La troisième lois est utile pour décrire les interactions entre système, mais c'est la deuxième qui permet d'obtenir l'équation vectorielle du mouvement.

- Exemple 1 : 
    
    *Figure 8: système modèle, n forces sont exercées sur le mobile.*
    
- Le bilan des forces: rappel ressort 1 + rappel ressort 2 + amortisseur + poids + réactions support, ce qui permet d'écrire: $\ \overrightarrow{F}_{ressort_1} + \overrightarrow{F}_{ressort_2} \overrightarrow{F}_{amortisseur} + m\overrightarrow{g} + \overrightarrow{R}_{sol}= m \overrightarrow{\gamma} \tag{44}$ 
A noter : la réaction du sol $\ \overrightarrow{R}_{sol}$ n'est pas une force unique mais différentes forces agissant aux divers points de contact entre les roues du wagon et le support, ces forces sont à priori distinctes les unes des autres, la combinaison de toutes ces réactions donne $\ \overrightarrow{R}_{sol}$ qui s'oppose au poids.

- Une deuxième méthode permet d'obtenir une équation du mouvement en n'utilisant le concept d'énergie plutôt que celui de forces. Un système posséde une énergie mécanique somme de son énergie cinétique de translation et d'énergies potentielles, en l'absence de phénomènes de dissipation ou de force extérieure motrice, l'énergie mécanique est conservée. Par contre si il a de la dissipation et des forces motrices alors, le **théorème de l'énergie mécanique**: dans un référentiel galiléen, un corpuscule de masse m se déplacant entre les positions A et B la variation d'énergie mécanique est égale au travail des forces non conservatives et motrices $\ \Delta E_m= E_{m_B} - E_{m_A} = W_{F_{non \, cons.}}^{A \rightarrow B} + W_{F_{motrices}}^{A \rightarrow B} \tag{45}$
Cette expression n'est pas l'équation du mouvement, mais dérivée par rapport au temps cela permet d'établir une équation différentielle du mouvement identique à l'application de la seconde lois de Newton.

--------
- Mouvements de rotation: ne sont considérés que le mouvement d'un objet solide en rotation autour d'un axe fixe dans un référentiel galiléen. Comme pour les mouvements en translation, plusieurs méthodes permettent de trouver l'équation du mouvement, deux seront utilisés le théorème du moment cinétique ou la dérivation de l'énergie mécanique.

--------

- **Théorème du moment cinétique**: il s'agit de l'équivalent en rotation de la seconde loi de Newton. Le moment cinétique est définit en point fixe O par $\ \overrightarrow{L}_O = m \, \overrightarrow{OM} \land \overrightarrow{v} \tag{46}$ pour un objet quasi-ponctuel situé en M se déplaçant à la vitesse $\ \overrightarrow{v}$, le moment cinétique est l'équivalent pour les mouvements de rotation de l'impulsion ou quantité de mouvement en translation.
    - Enoncé: La variation du moment cinétique est égale à la somme des moments des forces extérieures: $\ \frac{d \overrightarrow{L}_O}{dt} = \sum{\overrightarrow{M}_{F_{ext.}}} \tag{46}$ où le moment d'une force extérieure $\ \overrightarrow{F}$ s'appliquant sur la masse ponctuelle est: $\ \overrightarrow{M}_{F} = \overrightarrow{OM} \land \overrightarrow{F} \tag{47}$
    


- Exemple 2: le pendule pesant

*Figure 9: Le pendule pesant.*

Application du théorème du moment cinétique: Le point fixe est le point d'ancrage O, la masse est conscentrée au point M, $\ \overrightarrow{OM}=l \, \overrightarrow{e}_{\rho} \tag{48}$
Deux forces s'exercent sur la masse, son poids et une tension qui évite la chute libre. Ces forces s'écrivent $\ \left \{ \begin{array}{r l c}
\overrightarrow{P} & = & m g \overrightarrow{e}_x \\
\overrightarrow{T} & = &- T \overrightarrow{e}\rho \end{array} \right . \tag{49}$
Le moment cinétique est $\ \overrightarrow{L}_O = l \, \overrightarrow{e}_{\rho} \land (m v \overrightarrow{e}_{\varphi}) \tag{50}$ soit $\ \overrightarrow{L}_O = m\,l\,v \, \overrightarrow{e}_{z}  \tag{51}$ 
Donc $\ \frac{\overrightarrow{L}_O}{dt} = l\, \overrightarrow{e}_{\rho} \land (-T  \overrightarrow{e}_{\rho}) + l\, \overrightarrow{e}_{\rho} \land (m\,g  \overrightarrow{e}_{x})\tag{52} $
Le moment de la tension est nul, on utilise $\ \overrightarrow{e}_{\rho} = \cos(\varphi)\overrightarrow{e}_{x} + \sin(\varphi)\overrightarrow{e}_{y} \tag{53}$pour obtenir $\ \frac{\overrightarrow{L}_O}{dt} = -m\,g\, l\, \sin(\varphi)\overrightarrow{e}_{z} \tag{54} $
Cette équation vectorielle n'a qu'une seule composante dans la direction Oz, on peut passer d'une équation vectorielle à l'équation scalaire :
$\ m \, l \frac{dv}{dt} = - m\,g\,l \sin(\varphi) \tag{55} $

- L'exemple précédent est simple car l'objet en mouvement est une masse ponctuelle, si cette simplification ne peut être utilisée, le moment cinétique est évalué sur tout l'objet (supposé indéformable). Cette opération est facilitée par la notion de moment d'inertie $\ J_{\Delta}$ définit par $\ \newcommand{\deriv}{\mathrm{d}} 
\ J_{\Delta}=\int \!\!\! \int \!\!\! \int d(M,\Delta) \,\rho  \,\deriv V   \tag{56}$ où $\ d(M,\Delta)$ est la distance entre un point de l'axe $\ \Delta$ et un point M du solide, en ce point de volume $\ \deriv V$ la masse volumique est $\ \rho$. Contrairement à la masse, le moment d'inertie est définit par rapport à un axe, en général le calcul du moment d'inertie est le plus simple en considérant que l'axe $\ \Delta$ passe par le centre de masse du solide, si ce point n'est pas le point fixe, le moment d'inertie est évalué sur l'axe de rotation passant par un point fixe, en utilisant le théorème de Huygens-Schteiner: $\ J_{O,\Delta} = J_{C,\Delta} + M \times d^2 \tag{57}$
Le moment cinétique étendu à l'objet tournant autour de l'axe de rotation $\ \Delta$ passant par le point fixe O est 
$\ \overrightarrow{L}_O = J_{O,\Delta} \, \omega \overrightarrow{e}_{\Delta} \tag{58}$
Le théorème du moment cinétique est alors pour cet objet:
 $\ \frac{d \overrightarrow{L}_O}{dt} = \sum{\overrightarrow{M}_{F_{ext.}}} \tag{59}$
 où tous les moments des forces s'exerçant sur l'objet sont évalués.

### 1 A 3 2 Electrocinétique

Des exemples d'oscillateurs électroniques seront utilisés, ceux ci ne seront constitués que de dipoles passifs, résistances, capacités, inductances, transformateurs et de générateurs de courants ou  de tensions idéaux, donc ne sont pas au programme les transistors, les ampli-op, et autres.

Les grandeurs caractéritiques pertinentes pour l'étude des circuits sont les intensités de courant dans les branches et les différentes de potentiel entre les noeuds. Rappelons donc quelques définitions 

- Circuit: des éléments associés entre eux par des conducteurs, les éléments présents dans un circuit électrique sont *passifs* ou *actifs*, les éléments actifs sont les sources d'énergie, générateur de courant, piles, batteries, amplificateurs. Les éléments passifs sont les conducteurs, les interrupteurs, les résistances, les condensateurs, les bobines par exemple.

- Noeud-Branche: une *branche* est une association en série de conducteurs (fils de cuivre, pistes sur circuit imprimé), de bornes (physiquement l'endroit ou les élements sont reliés les uns aux autres), et d'élements dipolaires (qui sont reliés par deux bornes au circuit: une résistance est un élément dipolaire, un transistor à 3 bornes n'est pas un dipôle). Un circuit peut être constitué d'une seule branche fermée sur elle même dans ce cas il s'agit d'une boucle ou bien être connectée à d'autres branches. Un *noeud* est une borne ou plus de deux branches sont concourantes.

- Intensité: Le courant électrique est le phénomène physique décrivant le déplacement de toutes charges électriques, le courant est souvent caractérisé par son intensité, il est quantifié par le débit de charge électrique à travers une section de conducteur ou bien par des variations de quantité de charges électriques par rapport au temps:
$\ i = \frac{\deriv q}{\deriv t} \tag{65}$
la dimension de cette grandeur est le coulomb par seconde mais l'ampère est utilisé comme unité SI secondaire: 1 ampère = 1 coulomb/1 seconde. L'intensité du courant électrique est une quantité algébrique, elle peut être négative (des charges positives se déplacent à l'opposé du sens choisis) ou positive (des charges positives se déplacent dans le sens choisis).

*Figure 12: le sens conventionnel du courant, des charges positives se déplacent de la borne + vers la borne - à l'extérieur de la source.*


- Tension: la tension est une différence de potentiel entre deux points d'un circuit, le potentiel électrique V(M) en un point M d'un circuit est l'énergie potentielle $\ E_p$ d'une charge électrique q en ce point $\ V(M) = \frac{E_p}{q}$. le tension enre deux bornes A et B est la différence de potentiel entre ces deux bornes: $\ U_{AB} = V_A - V_B$.

*Figure 13: tension et potentiels.*


- Convention générateur, récepteur: Avant d'évaluer les tensions et les intensités dans un circuit, il est nécessaire de choisir la convention utilisée, celle ci est arbitraire ( comme le choix de la direction des axes dans un problème de mécanique) mais la convention va permettre d'utiliser les relations tension/intensité correctes pour les dipoles.

*Figure 14: Convention récepteur (a), générateur (b).*

Si le dipôle de la figure 14 est une résistance R, la tension V est égale à R I, la puissance absorbée est P = U I dans la convention récepteur, V est égale à -R I et la puissance absorbée P = -U I.

- Lois de Kirchhoff: Les deux lois de Kirchhoff permettent de trouver un ensemble de relations entre les tensions et les intensités dans un circuit:

    Loi des noeuds: la somme des intensités de courant arrivant dans un noeud est nulle.

    Dans la somme les intensités sont algébriques, positives si orientées vers le noeud, négatives dans le cas opposé.
    
    Loi des mailles: la somme des tensions le long d'une maille orientée est nulle.

    Dans la somme les tensions sont algébriques, négatives si orientées dans la direction de parcours de la maille (cette direction est au choix, elle n'a pas de conséquences physiques), positives dans le cas opposé.
